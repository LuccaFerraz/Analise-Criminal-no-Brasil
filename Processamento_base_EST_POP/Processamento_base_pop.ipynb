```python
import pandas as pd

est_pop = pd.read_csv("estimativa_pop_indicadores_esp.csv", sep=';',
                      encoding='ISO-8859-1')
```

```python
est_pop.dropna(inplace=True)
est_pop[['ano', 'cod_ibge']] = est_pop[['ano', 'cod_ibge']].apply(pd.to_numeric, errors='coerce').astype(int)
est_pop.rename(columns={'ano': 'Ano',
                        'nome_mun': 'Municipio'}, inplace=True)
 
```
